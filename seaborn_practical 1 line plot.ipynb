# python seaborn tutorial


```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7,8,9]
temp = [34,24,56,12,5,43,13,17,19]
temp_df = pd.DataFrame({"days":days,"temp":temp})
```

```python
sns.lineplot(x = "days",y = "temp", data = temp_df, )
plt.show()
```

```python
tips_data = sns.load_dataset("tips")
tips_data
```

```python
tips_data.shape()
```

```python
tr_dr = sns.lineplot( x = "total_bill",y = "tip",data = tips_data)
tr_dr
```

```python
tr_dr = sns.lineplot( x = "size",y = "tip",data = tips_data)
tr_dr
```

```python
tr_dr = sns.lineplot( x = "size",y = "tip",data = tips_data,hue = "sex",style = "sex",markers = ["o","<"])
tr_dr
```

```python
tr_dr = sns.lineplot( x = "size",y = "tip",data = tips_data,hue = "sex",style = "sex",markers = ["o","<"])
plt.title("title plot", fontsize = 16)
plt.xlabel("size", fontsize = 15)
plt.ylabel("tip", fontsize = 15)
plt.show()
```
